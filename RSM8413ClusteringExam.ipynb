## Clustering

### K-means

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1,3], [2,3], [3,1], [4,2], [6,5], [8,1], [8,3], [9,3]])

"""
Functionaly implementation for K-means
"""
def kmeans(X, k=3, max_iters=1, tol=1e-4):
    # 1. Initialize centroids outside of loop
    centroids = np.array([[3,2], [3,3], [6,3]])
    for i in range(max_iters):
        # 2. Assign observations to nearest point
        distances = np.linalg.norm(X[:, np.newaxis] - centroids, axis=2)
        labels = np.argmin(distances, axis=1)
        # 3. Update the centroid by taking the mean of the assigned points
        new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])
        # 4. Check to see if centroid values have stopped moving and converged
        if np.all(np.linalg.norm(new_centroids - centroids, axis=1) < tol):
            break
        centroids = new_centroids
    return centroids, labels


centroids, labels = kmeans(X, k=3)
# Visuals
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X')
plt.title("K-Means")
plt.show()
```
